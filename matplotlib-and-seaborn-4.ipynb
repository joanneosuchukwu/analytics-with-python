# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib.plot to show plot - not commonly used anymore
```

```python
x = np.random.randint(2, 20, 10)
y = x ** 2
```

# Functional Method

```python
plt.plot(x, y)
```

```python
np.linspace(2,10,5) #generates an array of 5 random numbers between 2 and 10 linearly spaced
```

```python
x = np.linspace(2, 10, 10)
y = x ** 2
```

```python
plt.plot(x, y)
```

```python
plt.subplot(2, 1, 1) #nrows, ncols, index
plt.plot(x,y, 'r')
plt.subplot(2, 1, 2)
plt.plot(y,x,'g')
```

# Object Oriented Method

* gives more control over graph than the functional method

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, height, buttom and width
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes.plot(x,y)
ax1 = fig.add_axes([0.3, 0.4, 0.5, 0.6]) 
```

```python
fig = plt.figure(figsize=(3,2), dpi=100) #width, height
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='x vs y', color='green', lw=5)
ax.plot(y,x, label='y vs x', color='red', ls='--', marker='o')
ax.legend()
```
